```python
import cv2
from PIL import Image
```

```python
img = cv2.imread("clear_image.jpg", flags = cv2.IMREAD_GRAYSCALE)
img
```

```python
Image.fromarray(img).show()
```

```python
_, new_img =cv2.threshold(img, 170, 255, cv2.THRESH_BINARY)
Image.fromarray(new_img).show()
```

```python
img2 = cv2.imread('dark_image.jpg', flags = cv2.IMREAD_GRAYSCALE)
img2
```

```python
Image.fromarray(img2).show()
```

```python
_, new_img2 = cv2.threshold(img2, 174, 255, cv2.THRESH_BINARY)
Image.fromarray(new_img2).show()
```

```python
new_img2 = cv2.adaptiveThreshold(
    img2, 255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    61,
    20
)
Image.fromarray(new_img2).show()
```
